```python
import tcxparser
import os
import ggps
from pathlib import Path
import fitparse
import pandas as pd
```

```python
activities = []
```

```python
path = "data/strava/activities"
tcx_files = Path(path).glob("*.gpx")
activityIdToLatLong = {}
activityIdtoLatLongList = {}
for filePath in tcx_files:
    activityId = filePath.name.split(".")[0]
    try:
        activity_ggps = ggps.GpxHandler().parse(str(filePath))
        activityIdToLatLong[activityId] = {
            "lat": activity_ggps.trackpoints[0].values['latitudedegrees'],
            "lng": activity_ggps.trackpoints[0].values['longitudedegrees'],
        }

        allLatLongs = list(map(lambda point: (point.values['latitudedegrees'], point.values['longitudedegrees']), activity_ggps.trackpoints))
        allLatLongsFiltered = [elem for indx, elem in enumerate(allLatLongs) if indx % 10 == 0]
        activityIdtoLatLongList[activityId] = allLatLongsFiltered
    except Exception:
        print(Exception)
```

```python
activities = pd.read_csv("data/strava/activities.csv")
activities.head(5)
```

```python
import json
activities["lat"] = 0.0
activities["lng"] = 0.0
activities["coordinates"] = "0"

for index, row in activities.iterrows():
    try:
        row = row.copy()
        activities.loc[index, 'lat'] = activityIdToLatLong[str(row['Activity ID'])]['lat']
        activities.loc[index, 'lng'] = activityIdToLatLong[str(row['Activity ID'])]['lng']
        activities.loc[index, 'coordinates'] = json.dumps(activityIdtoLatLongList[str(row['Activity ID'])])
    except:
        print("nope")
activities.columns
```

```python
activities.to_csv('activities_with_gpx_list.csv', index = False)
activities.to_json('activities_with_gpx_list.json')
activities.head(5).to_json('activities_with_gpx_list_small.json')
```
